```python
def add_score(subject_score,student,subject,score):
    appended = {str(student):{subject:score}}
    if(str(student) in subject_score.keys()):
        subject_score[str(student)].update(appended[str(student)])
    else:
        subject_score.update(appended)
    return subject_score
def calc_average_score(subject_score):
    result = {}
    for key in subject_score.keys():
        avg = sum(subject_score[key].values())/len(subject_score[key])
        result.update({key:str("%.2f"%avg)})
    return result

print(add_score({},'65010001','python',50))
print(add_score({'65010001':{'python':50}},'65010001','python',60))

print(calc_average_score({'65010001':{'python':60, 'calculus':55}, 
                            '65010002':{'python':50, 'calculus':45}}))
```

```python
def is_plusone_dictionary(d):
    check = []
    for key in d.keys():
        if(key+1 == d[key]):
            check.append(True)
        else:
            check.append(False)
    result = True
    for i in range(len(check)):
        result = result and check[i]
    return result

print(is_plusone_dictionary({1:2, 3:4, 5:6, 7:8}))
```

```python
def char_count(str):
    result = {}
    for i in str:
        if(i in result.keys()):
            result[i] += 1
        else:
            result.update({i:1})
    return result
print(char_count('language'))
```
